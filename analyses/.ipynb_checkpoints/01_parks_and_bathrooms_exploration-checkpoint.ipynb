# 01: Initial data exploration for NYC park and bathroom data

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

## Visitors in NY state parks in NYC

```python
# CSV from https://data.ny.gov/Recreation/State-Park-Annual-Attendance-Figures-by-Facility-B/8f3n-xj78
# Left the filename as-is for convenience

all_data = pd.read_csv('State_Park_Annual_Attendance_Figures_by_Facility___Beginning_2003.csv')
all_data.head()
```

```python
# Get only the state parks in NYC in the past few years
df = all_data.loc[(all_data['OPRHP Region'] == 'New York City') & (all_data['Year'] > 2015)]
df['Facility'] = df['Facility'].str.slice(0,20)
df.head(10)
```

```python
# Some simple counts
print(df['Facility'].unique())
print(len(df['Facility'].unique())) # Number of facilities (= state parks in NYC) in the data
```

```python
# Some plotting
sns.catplot(
    data=df, kind='bar',
    x='Facility', y='Attendance', hue='Year',
    height=6, aspect=12/6, palette='husl'
)
plt.ticklabel_format(style='plain', axis='y', useLocale=True)
plt.xticks(rotation=80)
plt.yticks()
plt.xlabel(None)
plt.ylabel('Number of annual visitors')
plt.title('Number of annual visitors of NY state parks in NYC', size=14)
plt.show()
```

## Bathroom data exploration

```python
# CSV from https://data.cityofnewyork.us/Recreation/Directory-Of-Toilets-In-Public-Parks/hjae-yuav
# The CSV was last updated February 2021
df = pd.read_csv('Directory_Of_Toilets_In_Public_Parks.csv')
df.head()
```

```python
# Look at how many parks have more than 1 bathroom
df_grouped = df.groupby('Name')\
    .count()[['Location']]\
    .rename({'Location': 'Number of bathrooms'}, axis=1)\
    .reset_index()
df_grouped.groupby('Number of bathrooms').count()
```

```python
# Some simple counts
print('Number of bathrooms:', len(df))
print('Number of accessible bathrooms:', len(df.loc[df['Handicap Accessible'] == 'Yes']))
```

```python
# Number of bathrooms per borough
df.groupby('Borough').count()[['Name']].rename({'Name': 'Number of park bathrooms'}, axis=1)
```

```python
# Number of accessible bathrooms per borough
df.loc[df['Handicap Accessible'] == 'Yes']\
    .groupby('Borough')\
    .count()[['Name']]\
    .rename({'Name': 'Number of accessible park bathrooms'}, axis=1)
```
